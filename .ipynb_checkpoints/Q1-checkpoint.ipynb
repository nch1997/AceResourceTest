```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import qqplot
import matplotlib.gridspec as gridspec
import numpy as np
from scipy.stats import shapiro, ttest_ind, f_oneway
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('ingredient.csv')
df.describe()
```

```python
# conduct normality test for each ingredient using Shaprio-Wilk Test
# if p-value is > 0.05 then data is normally distributed/parametric

for ingredient in df:
    stat, p = shapiro(df[ingredient])
    if p > 0.05:
        print(f' p-value is {p}. Therefore, {ingredient} is Parametric')
    else:
        print(f'p-value is {p}. Therefore, {ingredient} is Non Parametric')
```

```python
# Visual Proof for Parametric/Non-Parametric. Histograms, Boxplots and QQplots

fig, axs = plt.subplots(3, 3, figsize=(20, 10))
sns.histplot(data=df, x="a", kde=True, color="skyblue", ax=axs[0, 0])
sns.histplot(data=df, x="b", kde=True, color="olive", ax=axs[0, 1])
sns.histplot(data=df, x="c", kde=True, color="gold", ax=axs[0, 2])
sns.histplot(data=df, x="d", kde=True, color="teal", ax=axs[1, 0])
sns.histplot(data=df, x="e", kde=True, color="skyblue", ax=axs[1, 1])
sns.histplot(data=df, x="f", kde=True, color="olive", ax=axs[1, 2])
sns.histplot(data=df, x="g", kde=True, color="gold", ax=axs[2, 0])
sns.histplot(data=df, x="h", kde=True, color="teal", ax=axs[2, 1])
sns.histplot(data=df, x="i", kde=True, color="skyblue", ax=axs[2, 2])
```

```python
fig, axs = plt.subplots(3, 3, figsize=(20, 10))
sns.boxplot(data=df, x="a", color="skyblue", ax=axs[0, 0])
sns.boxplot(data=df, x="b", color="olive", ax=axs[0, 1])
sns.boxplot(data=df, x="c", color="gold", ax=axs[0, 2])
sns.boxplot(data=df, x="d", color="teal", ax=axs[1, 0])
sns.boxplot(data=df, x="e", color="skyblue", ax=axs[1, 1])
sns.boxplot(data=df, x="f", color="olive", ax=axs[1, 2])
sns.boxplot(data=df, x="g", color="gold", ax=axs[2, 0])
sns.boxplot(data=df, x="h", color="teal", ax=axs[2, 1])
sns.boxplot(data=df, x="i", color="skyblue", ax=axs[2, 2])
```

```python
fig, axes = plt.subplots(ncols=3, nrows=3, sharex=True, figsize=(3*3, 3*3))
for k, ax in zip(df.columns, np.ravel(axes)):
    qqplot(df[k], line='s', ax=ax)
    ax.set_title(f'{k}')
plt.tight_layout()
```

The statistical test and plots above proved that the ingredients "a" to "i" are non-parametric. 
<br>
Although a,b,d,g,e looks normal from the histogram and qqplot they are skewed to one side. Hence, they are also non-parametric

```python
# correlation test
# using spearman correlation coefficient because data is non parametric
sns.heatmap(df.corr(method='spearman'), annot=True, cmap="YlGnBu")
```

Although our data are non-parametric, parametric tests can perform well with skewed and nonnormal distributions provided that sample size guideliens are satisfied:

1. 1-sample t test - sample > 20.
2. 2-sample t test - sample > 15.
3. one-way ANOVA - sample > 15.

Since we have 214 data points we can proceed with parametric tests

Anything with above aboslute(0.5) correlation score are related. 
<br>
g-a and f-b are related. Use 2-sample t test to check significance.

```python
ga_stat, ga_p = ttest_ind(df['g'], df['a'], equal_var=False)
fb_stat, fb_p = ttest_ind(df['f'], df['b'], equal_var=False)

print(f'p-value for g-a is {ga_p}. It is smaller then 0.05. Therefore there is significant corerlation between the two')
print(f'p-value for f-b is {fb_p}. It is smaller then 0.05. Therefore there is significant corerlation between the two')
```

Check if each ingredient is significantly different from one another using one-way ANOVA

```python
fvalue, pvalue = f_oneway(df['a'], df['b'], df['c'], df['d'],
                          df['e'], df['f'], df['g'], df['h'],
                          df['i'])
if pvalue < 0.05:
    print(f' p-value: {pvalue} is < 0.05. Therefore, there are significant differences among the ingredients')
else:
    print(f' p-value: {pvalue} is > 0.05. Therefore, there are no significant differences among the ingredients')
```

Determining number of distinctice clusters using Clustering analysis

```python
range_n_clusters = range(2,15)
silhouette_avg = []

for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters)
    kmeans.fit(df)
    cluster_labels = kmeans.labels_
    silhouette_avg.append(silhouette_score(df, cluster_labels))
    

plt.plot(range_n_clusters,silhouette_avg, 'x-')
ax = plt.xlabel('Values of K') 
ax = plt.ylabel('Silhouette score') 
ax = plt.title('Silhouette analysis For Optimal k')
```

Based on the silhouette analysis the optimal number of k is 3. Hence, there are 3 distinctive formulations present in the dataset.
